```python
from ultralytics import YOLO
import cv2  
import torch
import numpy as np
```

```python
print(torch.cuda.is_available())
print(torch.cuda.device_count())
print(torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

```python
model=YOLO("yolo11n-seg.pt")
```

```python
model.train(
    data="custom_data.yaml", 
    epochs=200,                # Set high, Early Stopping will stop it anyway
    imgsz=640,
    batch=16,
    device=0,                  # RTX 2050
    
    # --- EARLY STOPPING ---
    patience=20,               # If no improvement for 20 epochs, stop training
    save=True,                 # Save the best model
    
    # --- DATA AUGMENTATION (For bad camera quality) ---
    blur=0.4,                  # High blur to simulate out-of-focus camera
    noise=0.15,                # High noise for grainy sensors
    degrees=180,               # Full rotation (since labels can be diagonal)
    perspective=0.001,         # Simulate different camera angles
    flipud=0.5,                # Flip upside down
    fliplr=0.5,                # Flip left-right
    mosaic=1.0,                # Essential for small datasets
    mixup=0.1,                 # Blends images to prevent overfitting
    hsv_s=0.4,                 # Color saturation jitter
    hsv_v=0.4                  # Brightness jitter (for changing light)
)
```
