```python
import easyocr
import time
import re
```

```python
def extract_text(image_path):
    """
    Extract all text from an image using EasyOCR (English only).
    Returns a list of text strings found in the image.
    """
    # Initialize reader for English
    reader = easyocr.Reader(['en'], gpu=False)
    
    # Extract text (detail=0 returns only text strings)
    results = reader.readtext(image_path, detail=0)
    
    return results
```

```python
image_file = r"C:\Users\IksuraW\Downloads\New folder\ClothTagNetV2\PXL_20260127_051346453.MP.jpg"  # Replace with your image path
    
    # Get text as list
text_list = extract_text(image_file)
print("Extracted text (list):")
for text in text_list:
    print(f"- {text}")
```

 GPU RUN

```python
start_time = time.time()
use_cuda = True
text_list = extract_text(image_file) #, use_gpu=use_cuda use this to enable gpu
end_time = time.time()
elapsed_time = end_time - start_time
paragraph = " ".join(text_list)
paragraph = paragraph.lower()  # Convert to lowercase for uniformity
print("Extracted Text:")
print(paragraph)
print(f"\nTime consumed: {elapsed_time:.2f} seconds")
```
