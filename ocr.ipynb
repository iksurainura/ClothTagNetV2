```python
import easyocr
import time
import re
```

```python
def extract_text(image_path):
    start_time = time.time()
    reader=easyocr.Reader(['en'],gpu=True)
    text_list=reader.readtext(image_path,detail=0)
    end_time = time.time()
    elapsed_time = end_time - start_time
    paragraph = " ".join(text_list)
    paragraph = paragraph.lower()
    print(f"\nTime consumed: {elapsed_time:.2f} seconds")
    return paragraph
```

```python
image_file = r"C:\Users\IksuraW\Downloads\New folder\ClothTagNetV2\test_img\PXL_20260127_051346453.MP.jpg"  # Replace with your image path
    
    # Get text as list
extract_text(image_file)
```

```python
start_time = time.time()
use_cuda = True
text_list = extract_text(image_file) #, use_gpu=use_cuda use this to enable gpu
end_time = time.time()
elapsed_time = end_time - start_time
paragraph = " ".join(text_list)
paragraph = paragraph.lower()  # Convert to lowercase for uniformity
print("Extracted Text:")
print(paragraph)
print(f"\nTime consumed: {elapsed_time:.2f} seconds")
```
